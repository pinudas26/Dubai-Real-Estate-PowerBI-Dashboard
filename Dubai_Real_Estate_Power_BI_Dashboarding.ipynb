**PROJECT NAME: DUBAI REAL ESTATE INTELLIGENCE DASHBOARD**

**Project Objective**

In order to provide real estate investors, agents, and decision-makers with valuable information, this project offers an interactive Power BI dashboard that analyzes housing data in Dubai. The goal was to turn unstructured, aesthetically pleasing property data into a report that supports well-informed pricing and investment strategies.

Important characteristics like the property's location, cost, size, number of bedrooms and bathrooms, type, and developer information were all included in the dataset. Inconsistencies and missing values were addressed, data types were standardized, and new computed fields—like price per square foot, property age, and listing category—were developed to enhance the analysis during the data preparation stage.

The dashboard highlights essential performance indicators, including total listings, average property price, average size, and price per square foot. Visual elements such as maps, bar charts, line charts, and treemaps were used to explore pricing patterns across locations, property types, and developers. Interactive slicers allow users to filter data dynamically by criteria like property type, number of bedrooms, and price range, making the dashboard flexible and user-centric.

Important findings show that location and size have a big impact on property pricing, with premium areas fetching far higher costs per square foot. While mid-range listings account for the majority of transactions, high-end properties are clustered among certain developers. These results imply that investors can balance their portfolios between premium and mid-range segments and select high-demand regions to maximize returns.

All things considered, the dashboard shows how analytics and data visualization can simplify complicated real estate data, allowing for quicker and more intelligent decision-making in a cutthroat market.

**Key Steps**


*   **Data Import & Understanding**

    Key characteristics including price, size (sqft), bedrooms, bathrooms, location, and property type were included in the housing information that was loaded into Power BI. To comprehend data distribution and spot discrepancies, preliminary investigation was carried out.
*   **Data Cleaning & Preparation**

    Problems with data quality, including duplicates, missing numbers, and inaccurate data types, were fixed. Categorical variables (location, property type) were standardized for consistency, while fields like price and size were transformed into numerical representations.


*   **Data Transformation**

    Created calculated columns as Price per Sqft, Property Age, Listing Category (Budget / Mid / High-End)
*   **Dashboard Design & Visualization**

    KPI cards and graphics including bar charts, maps, line charts, and treemaps were used to create an interactive dashboard. Pricing trends across locales, property kinds, and developers were examined with the aid of these graphics.

*   **Interactivity & User Experience**

    To enable dynamic exploration, slicers and filters (bedrooms, location, price range, property type) were included. Usability and narrative were enhanced by tooltips and layout design.

*   **Insight Generation**

    Important trends were found, including a substantial link between size and price and higher pricing in upscale locales. Market segmentation showed that mid-range listings dominated in terms of volume.

*   **Business Recommendations**

    Strategic recommendations were made based on observations, such as concentrating on high-demand regions, focusing on mid-range purchasers for volume, and investing in properties with higher price per square foot growth potential.

**SCATTER PLOT**

A scatter plot shows the relationship between two numerical variables using dots.

**Scatter Plot of Price and SquareFeet**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("housing_price_dataset.csv")

sns.scatterplot(x='SquareFeet', y='Price', data=df)

plt.title("Price vs SquareFeet")
plt.xlabel("Square Feet")
plt.ylabel("Price")

plt.show()
```

***Insight:***
*   Strong positive relationship
*   Bigger houses → Higher prices

**Scatter Plot of Price and Bedrooms**

```python
sns.scatterplot(x='Bedrooms', y='Price', data=df)

plt.title("Price vs Bedrooms")
plt.xlabel("Bedrooms")
plt.ylabel("Price")

plt.show()
```

***Insights:***
*   Price generally increases with bedrooms
*   But not perfectly linear (other factors matter)

**Scatter Plot of Price and Bathrooms**

```python
sns.scatterplot(x='Bathrooms', y='Price', data=df)

plt.title("Price vs Bathrooms")
plt.xlabel("Bathrooms")
plt.ylabel("Price")

plt.show()
```

***Insights:***
*   More bathrooms → Higher price (moderate correlation)

**Advanced Scatter Plot - Add color by Neighborhood**

```python
sns.scatterplot(x='SquareFeet', y='Price', hue='Neighborhood', data=df)

plt.title("Price vs SquareFeet by Neighborhood")
plt.xlabel("Square Feet")
plt.ylabel("Price")

plt.show()
```

***Insights:***
*   Compare price trends across Urban, Suburb, Rural
*   Identify which area has higher-priced properties

**PIE CHART**

The percentage share of each category in a dataset is displayed using a pie chart.

**Pie Chart of Neighborhood Distribution**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("housing_price_dataset.csv")

# Count values
counts = df['Neighborhood'].value_counts()

# Pie chart
plt.pie(counts, labels=counts.index, autopct='%1.1f%%')
plt.title("Distribution of Properties by Neighborhood")

plt.show()
```

***Insights:***
*   Shows percentage share of: Urban, Suburb and Rural           
*   Helps identify:Which area has maximum properties and
Market distribution
